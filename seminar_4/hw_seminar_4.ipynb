```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('./kc_house_data.csv', sep=',', encoding='windows-1251')
df.head(1)
```

* 1.2. Изучите стоимости недвижимости

```python
plt.figure(figsize=(6,4))
plt.hist(df['price'])
plt.title('График распределения стоимости недвижимости')
plt.xlabel('Стоимость')
plt.ylabel('Количество домов')
plt.xticks(rotation = 45)
```

Большая часть домов находится в диапозоне стоимости до 1 500 000 у.е. Больше всего домов (около 18000) стоят до 900 000 у.е.

* 1.3. Изучите распределение квадратуры жилой

```python
plt.figure(figsize=(6,4))
plt.hist(df['sqft_living'])
plt.title('График распределения жилой площади')
plt.xlabel('Площадь дома (жилая), кв.м')
plt.ylabel('Количество домов')
plt.xticks(rotation = 45)
```

Из графика видно, что большинство продаваемых домов имеют жилую площадь до 4000 кв.м. Максимальное количество продаваемых домов  площадью до 2400 кв.м

* 1.4. Изучите распределение года постройки

```python
data = df['yr_built'].value_counts()
data
```

```python
plt.figure(figsize=(6,4))
plt.hist(df['yr_built'])
plt.title('График распределения домов по годам постройки')
plt.xlabel('Год постройки')
plt.ylabel('Количество домов')
plt.xticks(rotation = 45)
```

Больше всего в продаже домов постройки конца 2000 года. Примерно равномерно распределены дома по количеству в диапазоне говов 1947-2005. Меньше всего продаётся домов 1900-1947 годов постройки
